```python
# Operation on Tuples
tuple1 = (1, 2, 3, 4, 5)
```

```python
# Reverse of tuple
tuple1 = tuple1[::-1]
print(tuple1)
```

```python
# Accessing the value in tuple
access = tuple1.index(3)
print(access)
```

```python
tuple2 = (50,)
print(tuple2)
```

```python
# swaping two tuple
tuple3 = (22, 44)
tuple4 = (66, 88)
```

```python
tuple3, tuple4 = tuple4, tuple3
print("tuple3",tuple3)
print("tuple4",tuple4)
```
