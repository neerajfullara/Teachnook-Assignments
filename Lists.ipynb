```python
# Operation on Lists
# List 1
list1 = [100, 200, 300, 400, 500]
```

```python
# Reversing the list
list1.reverse()
print(list1)
```

```python
# List 2
list2 = [1000, 200, 300, 100, 10, 500, 50, 5]
```

```python
# Sorting the List
list2.sort()
print(list2)
```

```python
# Concatinating two list
# list1.append(list2)
listT = list1+list2
print(listT)
```

```python
listT.sort()
print(listT)
```

```python
# Concatenating two list by indexing
l1 = ["I", "Lov", "t", "co"]
l2 = ["'m", "eing", "o", "de"]
```

```python
# By indexing
l3 = [i+j for i, j in zip(l1,l2)]
print(l3)
```

```python
# Calculating the square of elements in list
l4 = [10, 20, 30, 40, 50]
```

```python
# Using for loop for Calculation.
result = []
for i in l4:
  result.append(i*i)

print(result)
```
