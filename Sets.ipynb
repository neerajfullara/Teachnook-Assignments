```python
# Adding two sets
print("This the set of colors")
```

```python
set1 = {"Yellow","Green", "Black"}
```

```python
print("New set of colors")
```

```python
set2 = {"Blue", "Red", "Orange"}
```

```python
print("Adding set1 to set2")
new_set = set1.union(set2)
```

```python
print("New set is", new_set)
```

```python
# Removing elements from set
new_set.remove("Orange")
print(new_set)
```

```python
new_set.remove("Red")
print(new_set)
```

```python
#Adding new items
new_set.add("Pink")
print(new_set)
```

```python
new_set.add("Violet")
print(new_set)
```
