```python
# Operation Function
# Creating Function
def person(name, age:int):
  print(name)
  print(age)
```

```python
name = input("Enter Name - ")
age = int(input("Enter age -"))
```

```python
# Function call
person(name, age)
```

```python
# Taking Variable number of arguments in function
def percentage(*args):
  sum = 0
  for i in args:
    sum += i
  avgr = sum/len(args)
  print("Average", avgr)
```

```python
percentage(3,2)
```

```python
# Function for calculating addition and subtraction
def calculate(a, b):
  add = a+b
  sub = a-b
  return add, sub
```

```python
res = calculate(30,40)
print(res)
```
